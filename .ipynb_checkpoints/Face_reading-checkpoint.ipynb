```python
# Imports PIL module 
import cv2
import numpy as np
import matplotlib.pyplot as plt
import mediapipe as mp 
```

```python
mp_face_mesh = mp.solutions.face_mesh
```

```python
def save_face(data, fname = './face_0.txt'): 
    recap_landmarks = np.zeros((478,3))
    print(len(recap_landmarks))
    for i,l in enumerate(data.landmark):
        recap_landmarks[i] = [l.x, l.y, l.z]

    np.savetxt(fname, recap_landmarks)
```

```python
from os import listdir
from os.path import isfile, join

mypath = 'C:/Users/Enzo.Magal/Documents/Enzo2021/VidTIMIT/fadg0/video/sa1'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
for filename in onlyfiles :
    print(filename)
    filepath = mypath + '/' + filename
    img = cv2.imread(filepath)

    img = cv2.flip(img, 1)
    cv2.imshow('image', img)

    with mp_face_mesh.FaceMesh(max_num_faces = 1, 
                           refine_landmarks = True, 
                           min_detection_confidence = 0.5, 
                           min_tracking_confidence = 0.5) as face_mesh: 

        results = face_mesh.process(img)

    if results.multi_face_landmarks: 
        save_face(results.multi_face_landmarks[0], fname = "C:/Users/Enzo.Magal/Documents/Enzo2021/VidTIMIT/fadg0/face_mesh/sa1/face_" + str(filename))        

cv2.destroyAllWindows()
```
